# Define Functions

### Define a domain
In order for us to work with any functions, we need to define a range of possible x values. Let's do that in the cell below.

```python
# Your code here!
domain = np.linspace(0,1000,1000) # What numpy function helps us create an array of values?
```

### Linear Function
Let's first start out with defining a linear function. Define a function which takes in two parameters, slope and y-intercept, and then returns the range of the linear function. Then, visualize the function on a plot.

Remember, the equation for a linear function is: $$f(x) = mx + b \\ \text{ where m is the slope and b is the y-intercept.} $$

```python
# First, define a function!
def linear_function(slope,y_intercept):
    x = domain # Define a local variable equal to the domain
    y = slope * x + y_intercept # The function should return y, which is the set of all possible y values given the inputs
    return y

# Now, let's define variables to pass into our function!

slope = -1  # Your code here
y_intercept = 100 # Your code here

# Now, let's call the function and assign it to a variable called range
range = linear_function(slope, y_intercept)

# Plot
plt.figure()
plt.plot(domain, range) # Your code here

# Plot horizontal and vertical axes
plt.axvline(0, linewidth=1.5,color='black') # Your code here
plt.axhline(0, linewidth=1.5,color='black') # Your code here

# Plot Labels
plt.xlabel("x") # Your code here
plt.ylabel("f(x)") # Your code here
plt.title("Linear Function") # Your code here

# Show plot
plt.grid(visible='both')
plt.legend()
plt.show()
```

### Non-Linear Functions
#### Quadratic Function
Now, let's try to create a quadratic function. Define a function which takes in three parameters, a, b, and c, and then returns the range of the quadratic function. Then, visualize the function on a plot.

Remember, the equation for a quadratic function is: $$f(x) = ax^2 + bx + c$$

```python
# Quadratic Function
def quadratic_functions(a,b,c):
    x = domain # Define a local variable equal to the domain
    y = (a * x**2) + (b * x) + c  # Your code goes here
    return y

# Your values here!
a = -2
b = 0
c = 20

range = quadratic_functions(a,b,c)

plt.figure()
plt.plot(domain, range,'--',color='r')
plt.ylim(-100,100)
plt.xlim(-10,10)
plt.axvline(x=0, color='black', linestyle='-', linewidth=1.5)
plt.axhline(y=0, color='black', linestyle='-', linewidth=1.5)
plt.grid(visible='both')
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title("Quadratic Function")
plt.legend()
plt.show()
```

Define an exponential function!

```python
# Exponential Function
def exponential_function(a,b):
    x = domain
    y = a * (b**x)
    return y

# Your values here!
a = -1
b = 2

range = exponential_function(a,b)

plt.figure()
plt.plot(domain, range, ".",color="orange")
plt.axvline(x=0, color='black', linestyle='-', linewidth=1.5)
plt.axhline(y=0, color='black', linestyle='-', linewidth=1.5)
plt.grid(visible='both')

plt.ylim(-1,1)
plt.xlim(-10,10)

plt.xlabel("x")
plt.ylabel('f(x)')
plt.title("Exponential Function")
plt.show()
```

Define a natural logarithmic function

```python
# Natural Log Function
def natural_log_function():
    x = domain
    y = np.log(x)
    return y

# Call the function and set it equal to a variable called range

range = natural_log_function()

# Plot it
plt.figure()
plt.plot(domain, range)
plt.axvline(x=0, color='black', linestyle='-', linewidth=1.5)
plt.axhline(y=0, color='black', linestyle='-', linewidth=1.5)
plt.grid(visible='both')
plt.ylim(0,8)
plt.xlim(0,100)
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title("Natural Log Function")
plt.legend()
plt.show()
```
