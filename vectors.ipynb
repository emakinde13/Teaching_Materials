# Vectors
In this notebook, you will get practice working with and plotting 2d vectors! Vectors are ordered lists of numbers with both direction and magnitude. When we have a lot of information or data (neural spikes, fMRI, EEG, behavioral data, etc), we can represent them using vectors. The field of linear algebra gives us a way to work with these vectors and perform operations on them. 

Vectors generally have the following notation: $$\vec{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \\ \vdots \\ x_n\end{bmatrix}$$

## Vector Function

```python
import numpy as np
import matplotlib.pyplot as plt
def random_vector_generator(mean=10,stdev=1,dim=1):
    """
    returns a 1d vector containing elements pseudorandomly sampled from a gaussian
    params:
    mean (int, float) : mean of gaussian. (default = 10)
    stdev (int, float) : standard deviation of gaussian (default = 1)
    dim (int) : length of vector (default = 1)
    """
    empty_vector = np.zeros(dim)
    for i in range(len(empty_vector)):
        empty_vector[i] = np.random.normal(mean,stdev)
    return empty_vector

randVec1, randVec2 = random_vector_generator(), random_vector_generator
plt.figure()
plt.quiver(0,0,randVec1, randVec2)
plt.show()

```
